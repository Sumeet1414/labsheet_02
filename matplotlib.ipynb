```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=[5,6,7,8,9]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line graph")
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[20,30,40,50]
name=["may", "Feb", "March", "April"]
Color=["black","SkyBlue","Yellow","Silver"]
value=[0.2,0,0.5,0]
plt.pie(x, labels=name, colors=Color,explode=value,shadow=True)
plt.show()
```

```python
import numpy as np
X=np.array(["A","B","C","D"])
Y=np.array([3,8,1,10])
Color=["Brown","Blue","Black","yellow"]
plt.bar(X,Y,color=Color)
plt.show()
```

```python
x=np.array(["A",'B','C','D'])
y=np.array([3,8,1,10])
mycolors=["b","g","y","orange"]
plt.barh(x,y,color=mycolors)
plt.show()
```

```python
import numpy as np
X=np.array(["A","B","C","D"])
Y=np.array([2,4,6,8])
Color=["black","Blue","pink","Brown"]
Size=[900,500,600,400]
plt.scatter(X,Y,color=Color,marker="*",edgecolor="Green",s=Size,alpha=0.8)
plt.title("scatter",fontsize=20)
plt.xlabel("X axis",fontsize=20)
plt.ylabel("Y axis",fontsize=20)
plt.show()
```
